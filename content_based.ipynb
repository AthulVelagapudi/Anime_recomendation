```python
#id
#name
#genres
#type
#studios
#overview
#rating

data = anime[['MAL_ID','Name','Genres','Type']]
```

```python
data = data.merge(anime_overview[['MAL_ID','sypnopsis']])
```

```python
data = data[~data['sypnopsis'].isnull()]
```

```python
data = data[data['Type'] != 'Unknown']
```

```python
data = data[data['Genres'] != 'Unknown']
```

```python
data['Genres'] = data['Genres'].apply(lambda x:x.split(','))
```

```python
data['sypnopsis'] = data['sypnopsis'].apply(lambda x:x.split())
```

```python
data
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
data['Genres'] = data['Genres'].apply(collapse)
data['Type'] = data['Type'].apply(collapse)
```

```python
data['tags'] = data['Genres'] + data['Type'] + data['sypnopsis']
```

```python
new =  data.drop(columns=['Genres','Type','sypnopsis'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: x.lower())
new.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df = new
new_df['tags'] = new['tags'].apply(stem)
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
vector.shape
```

```python
vector
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
similarity.shape
```

```python
new_df[new_df['Name'] == 'Trigun'].index[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse = True,key =lambda x :x[1])[1:6]
```

```python
def recommend(anime):
    x = new_df[new_df['Name'] == anime].index[0]
    distances = similarity[x]
    anime_list = sorted(list(enumerate(distances)), reverse = True,key =lambda x :x[1])[1:11]
    for i in anime_list:
        print(new_df.iloc[i[0]].Name)
```
